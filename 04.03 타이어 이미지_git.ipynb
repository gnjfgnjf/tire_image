```python
import os
from PIL import Image

# 이미지가 저장된 디렉토리 경로 설정
train_normal = 'C:/Users/gnjfg/Downloads/archive (4)/Tire Textures/training_data/normal'  # 실제 디렉토리 경로로 변경해야 합니다.
train_crack = 'C:/Users/gnjfg/Downloads/archive (4)/Tire Textures/training_data/cracked'
test_normal = 'C:/Users/gnjfg/Downloads/archive (4)/Tire Textures/testing_data/normal'
test_crack = 'C:/Users/gnjfg/Downloads/archive (4)/Tire Textures/testing_data/cracked'

# 디렉토리 안의 모든 파일 목록 불러오기
train_n = os.listdir(train_normal)
train_c = os.listdir(train_crack)
test_n = os.listdir(test_normal)
test_c = os.listdir(test_crack)

# 디렉토리 안의 이미지 파일 목록 추출하기
train_n_files = [f for f in train_n if f.endswith(('jpg', 'jpeg', 'png'))]
train_c_files = [f for f in train_c if f.endswith(('jpg', 'jpeg', 'png'))]
test_n_files = [f for f in test_n if f.endswith(('jpg', 'jpeg', 'png'))]
test_c_files = [f for f in test_c if f.endswith(('jpg', 'jpeg', 'png'))]
```

```python
import os
from PIL import Image

def get_image_sizes(file_list, directory):
    sizes = []
    for image_file in file_list:
        image_path = os.path.join(directory, image_file)
        with Image.open(image_path) as img:
            width, height = img.size
            sizes.append((width, height))
    return sizes

def print_statistics(file_list, directory):
    sizes = get_image_sizes(file_list, directory)
    min_size = min(sizes)
    max_size = max(sizes)
    median_size = sorted(sizes)[len(sizes) // 2]

    print("최소값:", min_size)
    print("최대값:", max_size)
    print("중앙값:", median_size)

# 각 데이터셋에 대한 통계 출력
print("Train Normal Images:")
print_statistics(train_n_files, train_normal)

print("Train Crack Images:")
print_statistics(train_c_files, train_crack)

print("Test Normal Images:")
print_statistics(test_n_files, test_normal)

print("Test Crack Images:")
print_statistics(test_c_files, test_crack)
```

```python

def resize_and_convert_to_grayscale(file_list, directory, target_size=(1500, 1500)):
    resized_images = []
    for image_file in file_list:
        image_path = os.path.join(directory, image_file)
        img = cv2.imread(image_path)
        if img is not None:
            # 이미지 크기 조정
            img_resized = cv2.resize(img, target_size)
            # 이미지를 흑백으로 변환
            img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
            # 변환된 이미지를 다시 저장
            cv2.imwrite(image_path, img_gray)
            print(f"{image_path} resized to {target_size} and converted to grayscale")
            resized_images.append((image_file, img_gray))
        else:
            print(f"Failed to read {image_path}")
    return resized_images

# 각 데이터셋의 이미지 크기 조정 및 흑백으로 변환
print("Resizing and Converting Train Normal Images to Grayscale...")
resized_train_normal_images = resize_and_convert_to_grayscale(train_n_files, train_normal)

print("Resizing and Converting Train Crack Images to Grayscale...")
resized_train_crack_images = resize_and_convert_to_grayscale(train_c_files, train_crack)

print("Resizing and Converting Test Normal Images to Grayscale...")
resized_test_normal_images = resize_and_convert_to_grayscale(test_n_files, test_normal)

print("Resizing and Converting Test Crack Images to Grayscale...")
resized_test_crack_images = resize_and_convert_to_grayscale(test_c_files, test_crack)

```

```python
import random

# 랜덤하게 이미지를 선택하여 출력하는 함수 정의
def display_random_image(images):
    image_file, img = random.choice(images)
    cv2.imshow(image_file, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 랜덤하게 이미지 출력
print("Displaying a Random Resized and Converted Train Normal Image...")
display_random_image(resized_train_normal_images)

print("Displaying a Random Resized and Converted Train Crack Image...")
display_random_image(resized_train_crack_images)

print("Displaying a Random Resized and Converted Test Normal Image...")
display_random_image(resized_test_normal_images)

print("Displaying a Random Resized and Converted Test Crack Image...")
display_random_image(resized_test_crack_images)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# 데이터 준비
train_images = []
train_labels = []

# 손상되지 않은 타이어 이미지
for image_file, img in resized_train_normal_images[:100]:  # 예시로 100개의 이미지만 사용
    train_images.append(img)
    train_labels.append(0)  # 손상되지 않은 타이어를 나타내는 클래스: 0

# 손상된 타이어 이미지
for image_file, img in resized_train_crack_images[:100]:  # 예시로 100개의 이미지만 사용
    train_images.append(img)
    train_labels.append(1)  # 손상된 타이어를 나타내는 클래스: 1

train_images = np.array(train_images)
train_labels = np.array(train_labels)

# 모델 정의
model = models.Sequential([
    layers.Conv2D(64, (11, 11), strides=(4, 4), activation='relu', input_shape=(1500, 1500, 1)),
    layers.MaxPooling2D((3, 3), strides=(2, 2)),
    layers.Conv2D(128, (5, 5), activation='relu'),
    layers.MaxPooling2D((3, 3), strides=(2, 2)),
    layers.Conv2D(192, (3, 3), activation='relu'),
    layers.Conv2D(192, (3, 3), activation='relu'),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((3, 3), strides=(2, 2)),
    layers.Flatten(),
    layers.Dense(2048, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(2048, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(2, activation='softmax')  # 출력층: 손상되지 않은 타이어와 손상된 타이어를 분류하는 이진 분류
])

# 모델 컴파일
model.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])

# 모델 훈련
model.fit(train_images, train_labels, epochs=5, batch_size=16)  # 에포크 수와 배치 크기 조정
```

```python
import numpy as np

# 데이터 준비
test_images = []
test_labels = []

# 손상되지 않은 타이어 이미지
for image_file, img in resized_test_normal_images:
    test_images.append(img)
    test_labels.append(0)  # 손상되지 않은 타이어를 나타내는 클래스: 0

# 손상된 타이어 이미지
for image_file, img in resized_test_crack_images:
    test_images.append(img)
    test_labels.append(1)  # 손상된 타이어를 나타내는 클래스: 1

test_images = np.array(test_images)
test_labels = np.array(test_labels)

# 모델에 테스트 데이터 전달하여 예측 수행
predictions = model.predict(test_images)

# 예측 결과 출력
for i in range(len(predictions)):
    if np.argmax(predictions[i]) == test_labels[i]:
        print(f"Correct prediction: {predictions[i]}")
    else:
        print(f"Incorrect prediction: {predictions[i]}")
```

```python
# 모델 평가
test_loss, test_accuracy = model.evaluate(test_images, test_labels)

print(f'Test Accuracy: {test_accuracy}')
```
